## Container with most water




### Brute force approach

- find area of all possible combination
- then compare the area and return largest area found (timecomplexisty o(n^2)

### using two pointer approach
input = [1,8,6,2,5,4,8,3,7]




- intialize max_area = 0 to keep track of largest variable found so far
- current_area = min(1,7)*(8-0) = 8
    - where inside min is value from left and right viz 1 and 7 then (8-0) is position from (right - left) (above is array of size 9)
- move the shortest line inward for left pointer its +1 and right pointer is -1
- repeat this until the pointers meet
- time complexity: o(n)
