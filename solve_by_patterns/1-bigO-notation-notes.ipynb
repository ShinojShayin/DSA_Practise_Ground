### Big O Notation Cheat Sheet

Big O measures how an algorithm's **runtime** or **space requirements** grow as the input size ($n$) increases. It focuses on the "worst-case" scenario.

---

### Time Complexity Hierarchy
*From most efficient to least efficient:*

| Complexity | Name | Description |
| :--- | :--- | :--- |
| **$O(1)$** | Constant | Time remains the same regardless of $n$. |
| **$O(\log n)$** | Logarithmic | Time increases slowly (often cuts data in half). |
| **$O(n)$** | Linear | Time increases at the same rate as $n$. |
| **$O(n \log n)$** | Linearithmic | Standard for efficient sorting algorithms. |
| **$O(n^2)$** | Quadratic | Time increases exponentially (nested loops). |
| **$O(2^n)$** | Exponential | Time doubles with each addition to $n$. |
| **$O(n!)$** | Factorial | Becomes impossible to run very quickly. |



---

### Efficiency Gap (The Step Count)
This shows how many operations are performed as $n$ grows:

| $n$ | $O(\log n)$ | $O(n)$ | $O(n \log n)$ | $O(n^2)$ |
| :--- | :--- | :--- | :--- | :--- |
| **10** | 3 | 10 | 33 | 100 |
| **100** | 7 | 100 | 664 | 10,000 |
| **1,000** | 10 | 1,000 | 9,965 | 1,000,000 |



---

### Quick Rules for Analysis
1. **Drop the Constants:** $O(2n)$ becomes $O(n)$. We only care about the growth rate.
2. **Drop Less Significant Terms:** $O(n^2 + n)$ becomes $O(n^2)$. The highest power wins.
3. **Loops:** A single loop is $O(n)$. Nested loops are $O(n^2)$.
4. **Divide & Conquer:** If you are splitting data in half (like Binary Search), it's likely $O(\log n)$.
